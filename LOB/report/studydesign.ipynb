## Required stats

| Requirement | Detail | Criteria Check |
| :--- | :--- | :--- |
| **Replication Runs** | Run the simulation with **at least five different initial random seeds** for **each of the 8 scenarios** ($8 \times 5 = 40$ total runs minimum). | Guideline: "at least five different random seeds." |
| **Point Estimation** | Compute the **Mean** (average) for every $\text{Y}$ variable across the five runs per scenario. | Guideline: "compute the final point estimation." |
| **Confidence Interval (CI)** | Compute the **95% Confidence Interval (CI)** ($\alpha=0.05$) for every mean. | **CRUCIAL:** "Missing calculation of confidence intervals" is a common problem to avoid. |
| **Comparison Method** | State that you used the **Common Random Numbers (CRN) method** across your 8 scenarios. | Criterion: "Not using the CRN method when comparing different methods or scenarios" is a major error. |
| **Data Report (Raw)** | In the **Analysis** section, report the collected statistics for **each individual run** ($\text{run}_1, \text{run}_2$, etc.) for all 8 scenarios. | Guideline: "For each run, report the collected statistics." |
| **Data Report (Aggregated)** | For each scenario, find the **average stats (Mean and CI)** across all runs and present them in a clear table. | Guideline: "For each scenario, find the average stats..." |
